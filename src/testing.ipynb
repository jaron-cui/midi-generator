```python
from musiclang import Note
import random

Chord = List[int]
QuarterNote = List[Chord | 'hold']

def generate_one_melody_for_chords(chords):
  note_sequence = [None] * sum(int(chord.duration) for chord in chords)
  # add keyframe notes at the beginning of each chord change
  beat = 0
  for chord in chords:
    note_sequence[beat] = [random.choice(chord.chord_pitches + [None])]
    beat += int(chord.duration)
    
def generate_note_group(chord_pitches: List[int], possible_pitches: List[int], duration: int, measure_length: int) -> List[QuarterNote]:
  # possible note group types: 1. run [1, inf]; 2. trill [2, 8]; 3. hold [1, 4]
  tone_sequence = []
  while len(tone_sequence) < duration:
    start_pitch = random.choice(chord_pitches)  # possibly increase number of octaves
    end_pitch = random.choice(chord_pitches)
    remaining_duration = duration - len(tone_sequence)
    pattern_type = random.choice(['run', 'trill', 'hold'])
    if pattern_type == 'run':
      beat_split = random.choice([1, 2, 3, 4])
      sub_duration = random.randint(1, min(4, remaining_duration))
      run_direction = random.choice([-1, 1])
      last_pitch_index = possible_pitches.index(start_pitch)
      end_pitch_index = possible_pitches.index(end_pitch)
      for quarter in range(sub_duration):
        beats_left = (duration - quarter) * beat_split
        if random.randint(1, 2) == 1:
          run_direction = random.choice([-1, 1])
        if abs(end_pitch_index - last_pitch_index) >= beats_left:
          run_direction = 1 if end_pitch_index > last_pitch_index else -1
        beats = []
        for i in range(beat_split):
          pitch_index = last_pitch_index + i * run_direction
          relative_pitch_index = pitch_index % len(possible_pitches)
          pitch_octave = pitch_index // len(possible_pitches)
          new_pitch = possible_pitches[relative_pitch_index] + 12 * pitch_octave
          beats.append([new_pitch])
        tone_sequence.append(beats)
        last_pitch_index = last_pitch_index + beat_split * run_direction
    elif pattern_type == 'trill':
      beat_split = random.choice([1, 2, 3, 4])
      max_sub_durations = {1: 4, 2: 2, 3: 2, 4: 2}
      sub_duration = random.randint(1, min(max_sub_durations[beat_split], remaining_duration))
      pitch1 = random.choice(possible_pitches)
      pitch2 = end_pitch if sub_duration == remaining_duration else random.choice(possible_pitches + [pitch1 + random.choice([-12, 12])])
      beats = []
      for beat in range(sub_duration * beat_split):
        beats.append([pitch1 if beat % 2 == 0 else pitch2])
      tone_sequence.append(beats)
    elif pattern_type == 'hold':
      sub_duration = random.randint(1, min(measure_length, duration))
      pitch = random.choice(chord_pitches)
      tone_sequence.append([[pitch]])
      tone_sequence += [['hold']] * (sub_duration - 1)
  return tone_sequence
    
    
generate_one_melody_for_chords(chords)
```

```python
from midiutil import MIDIFile

def convert_note_group_sequence_to_midi(left_hand: List[QuarterNote], right_hand: List[QuarterNote], save_path):
  right_track = 0
  left_track = 1
  channel = 0
  tempo = 100
  volume = 100
  midi_file = MIDIFile(2)
  midi_file.addTempo(right_track, 0, tempo)
  
  for hand in [right_hand, left_hand]:
    if hand == right_hand:
      track, octave, volume = right_track, 4, 100
    else:
      track, octave, volume = left_track, 3, 75
    for time, quarter in enumerate(hand):
      note_duration = 1 / len(quarter)
      for i, note in enumerate(quarter):
        if note == 'hold':
          continue
        note_time = time + note_duration * i
        print(note)
        for pitch in note:
          midi_file.addNote(track, channel, pitch + 12 * octave, note_time, note_duration, volume)
  
  with open(save_path, 'wb') as output_file:
    midi_file.writeFile(output_file)
    
```

```python
import typing
import itertools
import numpy
import math


def chunks(lst, n):
  """Yield successive n-sized chunks from lst."""
  for i in range(0, len(lst), n):
      yield lst[i:i + n]
      
      
def ordered_sample(lst, n):
  return [lst[i] for i in sorted(random.sample(range(len(lst)), n))]


def generate_left_hand(chord_progression: List[typing.Tuple[List[int], int]]) -> typing.List[QuarterNote]:
  # 1. arpeggios, 2. runs, 3. jumping chords, 4. repeat chords
  notes = []
  pattern_type = None
  quarter_split = None
  num_measures_for_pattern = 0
  for chord_pitches, chord_duration in chord_progression:
    chord_pitches = sorted(list(set(chord_pitches)))
    if num_measures_for_pattern == 0:
      pattern_type = 'arpeggio'#random.choice(['arpeggio', 'run', 'jump', 'chords'])
      quarter_split = random.choice([1, 2])#random.choice([1, 2, 3])
      num_measures_for_pattern = random.randint(1, 4)
    num_measures_for_pattern -= 1
    # TODO: add chorded arpeggios
    if pattern_type == 'arpeggio':
      # arpeggios with variable length notes
      # any number of notes available in the chord
      # how many notes available in the duration?
      # could arpeggiate like V, ^, /, \
      max_distinct_pitches = chord_pitches + [pitch + 12 for pitch in chord_pitches] + [min(chord_pitches) + 24]
      max_num_distinct_pitches = len(max_distinct_pitches)
      
      possible_chord_splits = [1]
      for divisor in [2, 3]:
        if chord_duration % divisor == 0:
          possible_chord_splits += list(range(divisor, chord_duration + 1, divisor))
      
      
      arpeggio_type = random.choice(['monotonic', 'cresting'])
      max_arpeggio_length = max_num_distinct_pitches if arpeggio_type == 'monotonic' else max_num_distinct_pitches * 2 - 2
      all_arpeggio_cycles = [quarter_split * chord_split for chord_split in possible_chord_splits]
      possible_arpeggio_cycles = [cycle for cycle in all_arpeggio_cycles if 3 <= cycle <= max_arpeggio_length]
      cycle_length = random.choice(possible_arpeggio_cycles)
      chord_split = cycle_length / quarter_split
      
      if arpeggio_type == 'monotonic':
        # prefer to stick within one octave if possible with arpeggio length
        if cycle_length > len(chord_pitches) + 1:
          select_from = max_distinct_pitches
        else:
          select_from = chord_pitches + [min(chord_pitches) + 12]
        select_from = list(reversed(select_from)) if random.random() < 0.5 else select_from
        arpeggio_notes = ordered_sample(select_from, cycle_length)
      else:
        # prefer to stick within one octave if possible with arpeggio length
        if cycle_length > len(chord_pitches) * 2:
          select_from = max_distinct_pitches
        else:
          select_from = chord_pitches + [min(chord_pitches) + 12]
        # randomize ascending or descending
        select_from = list(reversed(select_from)) if random.random() < 0.5 else select_from
        # print('calc', tolerance, select_from, cycle_length)
        pre_crest_count = random.randint(cycle_length // 2 + 1, min(cycle_length, len(select_from)))
        post_crest_count = cycle_length - pre_crest_count
        # print('build', cycle_length, pre_crest_count, len(chord_pitches))
        arpeggio_notes = ordered_sample(select_from, pre_crest_count)
        # print('down', select_from[:select_from.index(arpeggio_notes[-1])], post_crest_count)
        # print('be', arpeggio_notes, select_from.index(arpeggio_notes[-1]))
        print('last note:', arpeggio_notes[-1], 'notes', select_from, 'postcount:', post_crest_count, 'pre:', pre_crest_count, 'cyc', cycle_length)
        arpeggio_notes += ordered_sample(list(reversed(select_from[:select_from.index(arpeggio_notes[-1])])), post_crest_count)
      arpeggio_notes = [[note] for note in arpeggio_notes]
      # print('add', chord_duration, chord_split)
      num_repeats = chord_duration / chord_split
      assert math.floor(num_repeats) == num_repeats
      # print('adding arpeggio', arpeggio_notes)
      notes += list(chunks(arpeggio_notes, quarter_split)) * int(num_repeats)
  return notes
```

```python
# https://en.wikipedia.org/wiki/Musical_form

class Form:
  def __init__(self, sections: List['Section']):
    self.sections = sections


class Scale:
  class _Types:
    def __init__(self, scale_pitches: List[int]):
      self._pitches = tuple(scale_pitches)
      
    @property
    def pitches(self) -> typing.Tuple[int]:
      return self._pitches
    
    @property
    def fifth(self) -> typing.Tuple[int]:
      return Scale.select_chord_notes(self.pitches, [(1, '♮'), (3, '♮'), (5, '♮')])
    
    @property
    def seventh(self) -> typing.Tuple[int]:
      return Scale.select_chord_notes(self.pitches, [(1, '♮'), (3, '♮'), (5, '♮'), (7, '♭')])
      
  class _Major(_Types):
    def __init__(self):
      super().__init__([0, 2, 4, 5, 7, 9, 11])
      
    @property
    def major_seventh(self) -> typing.Tuple[int]:
      return Scale.select_chord_notes(self.pitches, [(1, '♮'), (3, '♮'), (5, '♮'), (7, '♮')])
      
  class _Minor(_Types):
    def __init__(self):
      super().__init__([0, 2, 3, 5, 7, 8, 10])
    
  class _Augmented(_Types):
    def __init__(self):
      super().__init__([0, 3, 4, 7, 8, 11])
      
    @property
    def seventh(self) -> typing.Tuple[int]:
      return Scale.select_chord_notes(self.pitches, [(1, '♮'), (3, '♮'), (5, '♮'), (6, '♭')])
      
  class _Diminished(_Types):
    def __init__(self):
      super().__init__([0, 2, 3, 5, 6, 8, 9, 11])
      
  major = _Major()
  minor = _Minor()
  augmented = _Augmented()
  diminished = _Diminished()
    
  @classmethod
  def select_chord_notes(cls, scale_pitches: typing.Sequence[int], indices: typing.Sequence[typing.Tuple[int, str]]) -> typing.Tuple[int]:
    accidentals = {'♮': 0, '♭': -1, '♯': 1}
    chord_notes = [scale_pitches[i - 1] + accidentals[accidental] for i, accidental in indices]
    return tuple(chord_notes)
  
    
class Chord:
  def __init__(self, chord_pitches: List[int], scale_pitches: List[int], duration: int):
    self._chord_pitches = sorted(list(set(chord_pitches)))
    self._scale_pitches = sorted(list(set(scale_pitches)))
    self.duration = duration
    
  @property
  def unique_chord_pitches(self):
    return self._chord_pitches.copy()
    
  @property
  def octave_chord_pitches(self, octaves: int = 1):
    pitches = []
    for o in range(octaves):
      pitches += [pitch + o * 12 for pitch in self._chord_pitches]
    return pitches + [pitches[-1] + 12]
  
  @property
  def unique_scale_pitches(self):
    return self._scale_pitches.copy()
  
  @property
  def octave_scale_pitches(self, octaves: int = 1):
    pitches = []
    for o in range(octaves):
      pitches += [pitch + o * 12 for pitch in self._scale_pitches]
    return pitches + [pitches[-1] + 12]
  
def chords_indexing(chord_progression: List[Chord]) -> List[int]:
  seen_chords = {}
  num_unique_chords = 0
  indices = []
  for chord in chord_progression:
    if chord not in seen_chords:
      indices.append(seen_chords[chord])
    else:
      indices.append(num_unique_chords)
      seen_chords[chord] = num_unique_chords
      num_unique_chords += 1
  return indices

def generate_chord_progression(num_measures: int, measure_duration: int = 4) -> List[Chord]:
  all_chords = [
    Scale.major.fifth,
    Scale.major.seventh,
    Scale.major.major_seventh,
    Scale.minor.fifth,
    Scale.minor.seventh,
    Scale.augmented.fifth,
    Scale.diminished.fifth
  ]

# melodies will have like an 'up' part followed by a variation on the 'down' parts to be unresolved or 
class Phrase:
  def __init__(self, chord_progression: List[Chord], measure_duration: int = 4):
    self.chord_progression = chord_progression
    self.measure_duration = measure_duration
  
  
  
class Section:
  def __init__(self, components, left_config: 'SectionConfig', right_config: 'SectionConfig'):
    self.components = components
    self.left_config = left_config
    self.right_config = right_config
    self._variations = {}
  
  def to_notes(self, chord_progression, measure_duration: int):
    pass
  
  def _create_variation(self, left_reconfig: 'SectionConfig', right_reconfig: 'SectionConfig'):
    # TODO: make this actually create variation
    return Section(self.components, left_reconfig, right_reconfig)
  
  def make_variation(self, label: str, left_reconfig: 'SectionConfig', right_reconfig: 'SectionConfig') -> 'Section':
    if label not in self._variations:
      self._variations[label] = self._create_variation(left_reconfig, right_reconfig)
    return self._variations[label]
    
class SectionConfig:
  def __init__(self, transpose_by: int = 0, harmony_level: int = 1, resolve_ending: bool = False, ):
    self.transpose_by = transpose_by
    self.harmony_level = harmony_level
    self.resolve_ending = resolve_ending
```
